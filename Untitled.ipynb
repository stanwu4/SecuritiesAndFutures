```python
#performing t-test on buy sell, p value is 0, meaning there is a significant difference
buy_values = df.loc[df['direction'] == 'BUY','Total Latency']
sell_values = df.loc[df['direction'] == 'SELL','Total Latency']

t_statistic, p_value = ttest_ind(buy_values, sell_values)
print('t-statistic',t_statistic)
print('p-value',p_value)
```

```python
df['exchangeRspLatency'].mean()
lower_percentile = 0.05
upper_percentile = 0.95
grouped = df.groupby('direction')
filtered_data = []
for _, group in grouped:
    lower_bound = group['exchangeRspLatency'].quantile(lower_percentile)
    upper_bound = group['exchangeRspLatency'].quantile(upper_percentile)
    filtered_group = group[(group['exchangeRspLatency'] >= lower_bound) & (group['exchangeRspLatency'] <= upper_bound)]
    filtered_data.append(filtered_group)
filtered_df = pandas.concat(filtered_data)

# Plot the box and whisker plot without outliers using seaborn
sns.boxplot(data=filtered_df, x='direction', y='exchangeRspLatency')
```
